Write a Python program to perform the following tasks1.Create a pandas series from the below dictionary where indices are subjects:

{'English':{'Sam':60,'Jackson':74,'Ahree':85},
'History':{'Gloria':83,'Sam':65,'Isla':78,'Aron':72,'Gray':61},
'Geography':{'Jackson':92,'Gloria':95,'Isla':82,'Aron':75,'Ahree':76},
'Mathematics':{'Sam':99,'Gloria':74,'Jackson':89,'Ahree':85,'Gray':95},
'Science':{'Sam':89,'Aron':82,'Gray':78,'Isla':93,'Ahree':87}}

2.Convert the created series into DataFrame and replace the null values with zeroes.
3.Transpose the DataFrame and create a new column 'Average' and fill the values in it by
calculating the average of all subjects

```python
import pandas as pd
S = {'English':{'Sam':60,'Jackson':74,'Ahree':85}, 'History':{'Gloria':83,'Sam':65,'Isla':78,'Aron':72,'Gray':61}, 'Geography':{'Jackson':92,'Gloria':95,'Isla':82,'Aron':75,'Ahree':76}, 'Mathematics':{'Sam':99,'Gloria':74,'Jackson':89,'Ahree':85,'Gray':95}, 'Science':{'Sam':89,'Aron':82,'Gray':78,'Isla':93,'Ahree':87}}
S = pd.Series(S,index =['English', 'History', 'Geography','Mathematics','Science'])
S
```

```python
d = {'English':pd.Series([60,74,85], index= ['Sam','Jackson','Ahree']),
     'History':pd.Series([83,65,78,72,61], index= ['Gloria','Sam','Isla','Aron','Gray']),
'Geography': pd.Series([92,95,82,75,76], index =['Jackson','Gloria','Isla','Aron','Ahree']),
'Mathematics':pd.Series([99,74,89,85,95], index = ['Sam','Gloria','Jackson','Ahree','Gray']),
'Science':pd.Series([89,82,78,93,87],index= ['Sam','Aron','Gray','Isla','Ahree'])
}
df= pd.DataFrame(d)
df
```

```python
df.isnull().sum()
```

```python
df = df.fillna(0)
```

```python
df.isnull().sum()
```
